## Question 4

```python
salaries_data = pd.read_csv('../data/Salaries.csv', index_col='playerID', header=0, skiprows=[1])
salaries_data
```

```python
salaries_data[(salaries_data['teamID']=="ATL") & (salaries_data['salary']>1000000) | 
              (salaries_data['teamID']=="HOU") & (salaries_data['salary']>1000000)].index
```

```python
salaries_data[(salaries_data['teamID']=="ATL")].salary.describe()
```

```python
sala_dict = {"playerID":[], "yearID":[], "teamID":[], "lgID":[], "salary":[]}
for row in salaries_data.iterrows():
    sala_dict['playerID'].append(row[0])
    sala_dict['yearID'].append(row[1].yearID)
    sala_dict['teamID'].append(row[1].teamID)
    sala_dict['lgID'].append(row[1].lgID)
    sala_dict['salary'].append(row[1].salary)
```

```python
df = pd.DataFrame(data=sala_dict)
new_col = {'playerID':'a', 'yearID':'b', 'teamID':'c', 'lgID':'d', 'salary':'e'}
df.rename(columns=new_col, inplace=True)
df
```

```python
x = np.linspace(-5, 5, num=40)
y = np.array([- j ** 2 - j + 3 for j in x])

plt.plot(x,y)
plt.title("Curve")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
x = np.linspace(-5, 5, num=40)
y = np.array([- j ** 2 - j + 3 for j in x])
z = np.array([- 0.5 * j ** 2 - j + 5 for j in x])


plt.plot(x,y, label='y')
plt.plot(x,z, label='z')
plt.legend(loc="best")
plt.text(-5, 2, '$y=-0.5*x^2-x+5$', fontsize=8)
plt.text(-5, -10, '$y=-x^2-x+3$', fontsize=8)
plt.title("Compare")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = np.linspace(1, 20, num=80)
y1 = np.array([j for j in x])
y2 = np.array([j ** 0.5 for j in x])
y3 = np.array([np.log(j) for j in x])

fig,axes=plt.subplots(nrows=3,ncols=1,figsize=(10,10))
print(fig)
print(axes)


fig.suptitle("x vs sqrt(x) vs logx",fontsize=15)

axes[0].set_title("Sub plot 1: x vs y1",fontsize=15,fontstyle='oblique')
axes[0].plot(x,y1)


axes[1].set_title("Sub plot 2: x vs y2",fontsize=10,fontstyle='normal')
axes[1].plot(x,y2)


axes[2].set_title("Sub plot 3: x vs y3",fontsize=5,fontstyle='italic')
axes[2].plot(x,y3)
```

```python
x = np.linspace(1, 10000, num=10000)
y = np.array([np.log(j) for j in x])


plt.xscale('log')
plt.yscale('log')
plt.ylim(ymin=1,ymax=1000)
plt.xlim(xmin=1,xmax=100000)

plt.plot(x,y)
plt.show()
```
